```
dbutils.fs.rm("/tmp/episodes.out.avro", True);
dbutils.fs.rm("/tmp/movies.out.avro", True);
```

```
# Create a DataFrame from a specified directory
df = spark.read.format("avro").load("/FileStore/tables/episodes.avro")
print("DF:", df)

#  Saves the subset of the Avro records read in
subset = df.where("doctor > 5")
print("subset:", subset)
subset.show()

subset.write.format("avro").save("/tmp/episodes.out.avro")
```

```
%scala

val df = Seq((2012, 8, "Batman", 9.8),
    (2012, 8, "Hero", 8.7),
    (2012, 7, "Robot", 5.5),
    (2011, 7, "Git", 2.0))
    .toDF("year", "month", "title", "rating")

df.write.mode("overwrite").partitionBy("year", "month").format("avro").save("/tmp/movies.out.avro")

display(dbutils.fs.ls("/tmp/movies.out.avro"))

val data = spark.read.format("avro").load("/tmp/movies.out.avro")

display(data)
```

```
%sql
CREATE TEMPORARY VIEW avroTable USING avro OPTIONS (path "/tmp/movies.out.avro");

SELECT * FROM avroTable;
```
